```python
network = UNet(in_nc=3,
               out_nc=3,
               n_feature=48)
network.to("cuda")

est_stack = np.zeros([91, img.shape[0], img.shape[2]])
with tqdm(total=91) as pbar:
    for model_id in range(90):
        network.load_state_dict(torch.load("results/unet_gauss25_b4e100r02/2022-04-20-20-45/epoch_model_0{}.pth".format(model_id+10)))
        network.eval()

        # for i in range(img.shape[1]):
        for i in [380, ]:
            test_in = np.repeat(img[:, [i], :], 3, 1)
            test_in = torch.from_numpy(test_in.transpose(1,0,2)).unsqueeze(0)
            with torch.no_grad():
                test_out = network(test_in.float().cuda()/255)
        est_stack[model_id] = test_out.detach().cpu().numpy().squeeze()[0]
        pbar.update(1)
```

```python
import torchio as tio
import torch
import skimage.io as sio
from tqdm import tqdm
import matplotlib.pyplot as plt

patch_loader_list = []
aggregator_list = []

subject = tio.Subject(one_image = tio.ScalarImage('../Documents/data/OCT-Scans/0P7_OS_V_3x3_0_0000059.tif'))
patch_overlap = 2
p_size = 16

grid_sampler = tio.inference.GridSampler(subject, p_size, patch_overlap)
patch_loader_list.append(torch.utils.data.DataLoader(grid_sampler, batch_size=16, shuffle=False))
aggregator_list.append(tio.inference.GridAggregator(grid_sampler, overlap_mode='average'))

s = 0
patch_loader = patch_loader_list[s]
aggregator = aggregator_list[s]
with tqdm(total=len(patch_loader)) as pbar:
    for batch, data in enumerate(patch_loader):
        locations = data[tio.LOCATION]
        aggregator.add_batch(data['one_image']['data'], locations)
        pbar.update(1)

data_recon = aggregator.get_output_tensor()
        
plt.figure(figsize=(10, 10))
plt.imshow(data_recon[0, :, 250].detach().cpu().numpy())

plt.figure(figsize=(10, 10))
plt.imshow(subject.one_image.data[0, :, 250].detach().cpu().numpy())

plt.figure(figsize=(10, 10))
plt.imshow(subject.one_image.data[0, :, 250].detach().cpu().numpy()-data_recon[0, :, 250].detach().cpu().numpy())

plt.show()
```
